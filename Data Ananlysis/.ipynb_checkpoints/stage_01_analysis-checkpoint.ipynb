```python
import pandas as pd
import numpy as np

purchase_data = pd.read_csv('../Data/QVI_purchase_behaviour.csv')
transaction_data = pd.read_excel('../Data/QVI_transaction_data.xlsx')
```

```python
merged_data = pd.merge(purchase_data, transaction_data, on='LYLTY_CARD_NBR', how='outer')
```

```python
merged_data.head(5)
```

```python
merged_data.info()
```

```python
merged_data.isnull().sum()
```

```python
# Calculate Q1, Q3, and IQR
Q1 = merged_data['TOT_SALES'].quantile(0.25)
Q3 = merged_data['TOT_SALES'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = merged_data[(merged_data['TOT_SALES'] < lower_bound) | (merged_data['TOT_SALES'] > upper_bound)]
print("Outliers:\n", outliers)
```

```python
# Separate numerical columns
numerical_data = merged_data.select_dtypes(include=['int64', 'float64'])

# Separate categorical columns
categorical_data = merged_data.select_dtypes(include=['object', 'category'])

# Print results
print("Numerical Columns:")
print(list(numerical_data.columns))

print("\nCategorical Columns:")
print(list(categorical_data.columns))
```

```python
# total sales
total_sales = merged_data['TOT_SALES'].sum()
print("Total Sales:", total_sales)
```

```python
# Sales by Product or Category
sales_by_category = merged_data.groupby('PROD_NAME')['TOT_SALES'].sum()
print(sales_by_category)
```

```python
# Loop through categorical columns and display unique values
categorical_columns = ['LIFESTAGE', 'PREMIUM_CUSTOMER', 'PROD_NAME']  # Replace with your column names

for col in categorical_columns:
    print(f"Unique values in {col}:")
    print(merged_data[col].unique())  # Displays unique values
    print("\n")
```

```python
import matplotlib.pyplot as plt

# Group by 'LIFESTAGE' and calculate total sales
sales_by_lifestage = merged_data.groupby('LIFESTAGE')['TOT_SALES'].sum()

# Print the total sales by lifestage
print("Total Sales by Lifestage:")
print(sales_by_lifestage)
```

```python
# Visualize the results using a bar chart
sales_by_lifestage.plot(kind='bar', color='skyblue', figsize=(8, 4))
plt.title("Total Sales by Lifestage")
plt.ylabel("Total Sales")
plt.xlabel("Lifestage")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by PREMIUM_CUSTOMER and calculate total sales
sales_by_premium = merged_data.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].sum()
print("\nSales by Premium Customer:")
print(sales_by_premium)
```

```python
sales_by_premium.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Sales Distribution by Premium Customer Status")
plt.ylabel("")
plt.show()
```

#### ***Converting excel series date to DateTime format***

```python
# Converting the excel series date to datetime
merged_data['DATE'] = pd.to_datetime(merged_data['DATE'], origin='1899-12-30', unit='D')

# Print the first few rows to verify
print(merged_data['DATE'].head())
```

#### ***Aggregate sales by month***

```python
# Extract month and calculate total sales by month
merged_data['Month'] = merged_data['DATE'].dt.month
sales_by_month = merged_data.groupby('Month')['TOT_SALES'].sum()

print("Sales by Month:")
print(sales_by_month)
```

#### ***Aggregate Sales by Day***

```python
# Extract day of the week and calculate total sales
merged_data['Day'] = merged_data['DATE'].dt.day_name()  # Get names like 'Monday', 'Tuesday', etc.
sales_by_day = merged_data.groupby('Day')['TOT_SALES'].sum()

print("Sales by Day of Week:")
print(sales_by_day)
```

#### ***Visualizing Monthly Trends***

```python
sales_by_month.plot(kind='line', marker='o', color='blue', figsize=(8, 4))
plt.title("Sales Trends by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13))  # Ensure all months are labeled
plt.grid(True)
plt.show()
```

#### ***Visualizing Weekly Trends***

```python
sales_by_day.plot(kind='bar', color='skyblue', figsize=(8, 4))
plt.title("Sales Trends by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)  # Rotate for readability
plt.show()
```

#### ***Find Patterns Over Time***

```python
monthly_segment_sales = merged_data.groupby(['LIFESTAGE', 'Month'])['TOT_SALES'].sum().unstack('LIFESTAGE')
monthly_segment_sales.plot(kind='line', figsize=(10, 4))
plt.title("Monthly Sales Trends by Lifestage")
plt.xlabel("Month")
plt.ylabel("Total Sales")
# plt.legend(title="Lifestage")
plt.show()
```
